```python
center1 = tuple(map(int, results[1].keypoints.data[0][0]))
center2 = tuple(map(int, results[1].keypoints.data[0][1]))
center3 = tuple(map(int, results[1].keypoints.data[0][2]))
center4 = tuple(map(int, results[1].keypoints.data[0][3]))

# Drawing a filled circle
cv2.circle(img, center1, 10, (0, 0, 255), -1)
cv2.circle(img, center2, 10, (0, 0, 255), -1)
cv2.circle(img, center3, 10, (0, 0, 255), -1)
cv2.circle(img, center4, 10, (0, 0, 255), -1)
```
