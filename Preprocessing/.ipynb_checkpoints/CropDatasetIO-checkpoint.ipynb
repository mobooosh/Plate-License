```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
from IPython.display import clear_output
import shutil
import csv
import random
```

```python
start_point = (0,0)
points = []
drawing= False

def create_pts1(points):
    pts1 = []
    for j in range(len(points)):
        temp = []
        for x, y in points[j]:
            temp.append(x)
            temp.append(y)
        pts1.append(temp)
    pts1 = np.array(pts1, dtype='float32')
    return pts1
    
def annotate_image_with_undo(event, x, y, flags, param):
    global points, drawing

    if event == cv2.EVENT_LBUTTONDOWN:
        start_point = (x, y)
        drawing = True 
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing:
            temp_image = image.copy()
            cv2.imshow("image", temp_image)
    elif event == cv2.EVENT_LBUTTONUP:

        points.append([(x, y)])
        cv2.circle(image, (x, y), 4, (0, 0, 255), thickness = -1)
        cv2.imshow("image", image)
        drawing = False



for image_name in os.listdir('data'):
    flag_end = False
    flag_delete = False
    points = []
    random_number = random.randint(1000, 1000000)
    path_img = 'data/' + image_name
    image = cv2.imread(path_img)
    clone = image.copy()
    temp_image = image.copy()
    cv2.namedWindow("image")
    cv2.setMouseCallback("image", annotate_image_with_undo)
    
    
    while True:
        if drawing == False:
            cv2.imshow("image", image)
        key = cv2.waitKey(5) & 0xFF

        if key == ord("r"):
            image = clone.copy()
            points = []
        
        elif key == ord("z"):
            image = clone.copy()
            points.pop()
            for p in points:
                cv2.circle(image, p[0], 4, (0, 0, 255), thickness = -1)

        elif key == 27:
            break

        elif key == ord("q"):
            flag_end = True
            break
            
        elif key == ord("d"):
            flag_delete = True
            break
            
    if flag_end == True: 
        break


    if flag_delete == True:
        os.remove(path_img)
        continue
    
    name_of_plate = input('Enter Plate Number: ')
    image_path = 'newData/' + name_of_plate + '_' + str(random_number) + '.jpg'
    clear_output(wait=True)

    cv2.destroyAllWindows()

    pts1 = create_pts1(points)

    pts2 = np.float32([[0,0],[256,0],[0,65],[256,65]])
    if len(pts1) == 4:
        M = cv2.getPerspectiveTransform(pts1, pts2)
        dst = cv2.warpPerspective(clone, M, (256, 65))
        cv2.imwrite(image_path, dst)
        shutil.move(path_img, 'cropped_image/' + name_of_plate + '_' + str(random_number) + '.jpg')

    text_file_name = 'cropped_image/' + name_of_plate + '_' + str(random_number) + '.csv'
    with open(text_file_name, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['Left-top', 'Right-top', 'Bottom-Left', 'Bottom-Right'])
        writer.writerow([pts1[0], pts1[1], pts1[2], pts1[3]])


cv2.destroyAllWindows()
```
