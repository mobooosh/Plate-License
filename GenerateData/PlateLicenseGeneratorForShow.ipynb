```python
import os
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

# Read Path

```python
template_plate_path = []

for template in os.listdir('syyd/template/'):
    template_plate_path.append(os.path.join('syyd/template/', template))
```

```python
word_path = []

for word in os.listdir('syyd/word/'):
    word_path.append(os.path.join('syyd/word/', word))
```

## Police Template

```python
test_path = template_plate_path[2]
temp = cv2.imread(test_path)
template = temp[45:390, 15:1553]
test_path = 'test-P.jpg'
word = cv2.imread(test_path)
cropped_image = word[378:635, 353:647]
resized = cv2.resize(cropped_image, (250, 210))
ret, resized = cv2.threshold(resized, 127, 255, cv2.THRESH_BINARY_INV)
logo_h, logo_w, _ = resized.shape
location_h = 110
location_w = 450
padded_img2 = np.zeros_like(template)
padded_img2[location_h:location_h+logo_h,location_w:location_w+logo_w,:] = resized

output_image = cv2.add(template, padded_img2)
output_image = output_image[...,::-1]
plt.imshow(output_image)
```

## white template

```python
test_path = template_plate_path[0]
temp = cv2.imread(test_path)
template = temp[45:390, 15:1553]
test_path = 'test-P.jpg'
word = cv2.imread(test_path)
cropped_image = word[378:635, 353:647]
resized = cv2.resize(cropped_image, (250, 210))
ret, resized = cv2.threshold(resized, 127, 255, cv2.THRESH_BINARY_INV)
logo_h, logo_w, _ = resized.shape
location_h = 110
location_w = 450
padded_img2 = np.zeros_like(template)
padded_img2[location_h:location_h+logo_h,location_w:location_w+logo_w,:] = resized
ret, padded_img2 = cv2.threshold(padded_img2, 127, 255, cv2.THRESH_BINARY)

output_image = cv2.bitwise_xor(template, padded_img2)
output_image = output_image[...,::-1]
plt.imshow(output_image)
```
