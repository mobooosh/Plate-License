```python
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
if 'validation' not in os.listdir('../Plate_license'):
    os.mkdir('validation')
if 'test' not in os.listdir('../Plate_license'):
    os.mkdir('test')
if 'train' not in os.listdir('../Plate_license'):
    os.mkdir('train')
```

```python
base_dir = './data'
list_data = os.listdir(base_dir)
```

```python
kernel = np.ones((3, 3), np.uint8)
```

```python
gray = cv2.cvtColor(example, cv2.COLOR_BGR2GRAY)
blur = cv2.blur(gray, (3, 3), 0)
threshold = cv2.adaptiveThreshold(blur, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
blur2 = cv2.blur(threshold, (3, 3), 0)
ret, thershold2 = cv2.threshold(blur2, 105, 255, cv2.THRESH_BINARY)
```

```python
plt.imshow(threshold, cmap='gray')
```

```python
plt.figure(figsize=[10, 5])
plt.subplot(231);plt.imshow(gray, cmap='gray');plt.title('gray')
plt.subplot(232);plt.imshow(blur, cmap='gray');plt.title('blur')
plt.subplot(233);plt.imshow(threshold, cmap='gray');plt.title('threshold')
plt.subplot(234);plt.imshow(blur2, cmap='gray');plt.title('blur')
plt.subplot(235);plt.imshow(thershold2, cmap='gray');plt.title('thershold2')
```

```python
import shutil
valid_and_test = 0.1
number_of_validation = int(len(list_data) * valid_and_test)
number_of_test = int(len(list_data) * valid_and_test)
number_of_train = len(list_data) - (number_of_test + number_of_validation)

for data in list_data:
    if number_of_validation != 0:
        number_of_validation -= 1
        # move data to validation directory
        shutil.move(os.path.join(base_dir, data), './validation')
    else:
        if number_of_test != 0:
            number_of_test -= 1
            # move data to test directory
            shutil.move(os.path.join(base_dir, data), './test')
        else:
            if number_of_train != 0:
                number_of_train -= 1
                # move data to train directory
                shutil.move(os.path.join(base_dir, data), './train')
```
